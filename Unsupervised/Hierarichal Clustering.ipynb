```python
X
```

```python
y=iris.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.33)
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled=pca.fit_transform(X_train_scaled)
```

## Agglomerative Clustering

**To Construct Dendogram**

```python
import scipy.cluster.hierarchy as sc

#plot the dendogram

plt.figure(figsize=(20,7))
plt.title("Dendogram")

## Create dendogram
sc.dendrogram(sc.linkage(X_train_scaled,method='ward'))
plt.xlabel('Sample Index')
plt.ylabel("Eucldeian Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg=AgglomerativeClustering(n_clusters=2)
agg.fit(X_train_scaled)
```

```python
agg.labels_
```

```python
plt.scatter(X_train_scaled[:,0],X_train_scaled[:,1],c=agg.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silo_coefficents=[]

for i in range(2,11):
    ag=AgglomerativeClustering(n_clusters=i)
    ag.fit(X_train_scaled)
    score=silhouette_score(X_train_scaled,ag.labels_)
    silo_coefficents.append(score)
```

```python
silo_coefficents
```

```python
plt.plot(range(2,11),silo_coefficents)
plt.xticks(range(2,11))
```

```python
plt.plot(range(2,11),silo_coefficents)
plt.xticks(range(2,11))
plt.xlabel("Number of cluster")
plt.ylabel("Silhotte Coeffecient")
plt.show()
```
